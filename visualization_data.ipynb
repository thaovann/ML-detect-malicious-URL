```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv("MaliciousURL.csv")

# Hiển thị thông tin cơ bản về dữ liệu
print(data.head())

# Đếm số lượng URL theo từng loại
url_count = data["type"].value_counts()

# Vẽ biểu đồ Bar chart
plt.figure(figsize=(10, 6))
url_count.plot(kind="bar", color="skyblue")
plt.title("Phân phối loại URL")
plt.xlabel("Loại URL")
plt.ylabel("Số lượng")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv("MaliciousURL.csv")

# Đếm số lượng URL theo từng loại
url_count = data["type"].value_counts()

# Màu sắc cho từng loại URL
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]  # Thay đổi màu sắc tại đây

# Vẽ biểu đồ Pie chart với màu sắc được thay đổi
plt.figure(figsize=(8, 8))
plt.pie(
    url_count, labels=url_count.index, autopct="%1.1f%%", startangle=140, colors=colors
)
plt.title("Phân phối loại URL", y=1.05)
plt.axis("equal")  # Để tạo hình tròn
plt.show()
```

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
data = pd.read_csv("MaliciousURL.csv")

# Đếm số lượng URL theo từng loại
url_count = data["type"].value_counts()

# Tổng số lượng các loại URL
total_urls = url_count.sum()

print(f"Tổng số lượng các loại URL: {total_urls}")
```
